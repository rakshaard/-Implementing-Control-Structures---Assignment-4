```python
# =========================================================
# Assignment 4: Implementing Control Structures
# Employee Schedule Management System
# Language: Python
# =========================================================

import random

# Days and shifts
days = ["Monday", "Tuesday", "Wednesday", "Thursday",
        "Friday", "Saturday", "Sunday"]

shifts = ["Morning", "Afternoon", "Evening"]

# ---------------------------------------------------------
# Employee Data
# Each employee gives preferred shift for each day
# ---------------------------------------------------------

employees = {
    "Alice": {
        "Monday": "Morning",
        "Tuesday": "Morning",
        "Wednesday": "Afternoon",
        "Thursday": "Evening",
        "Friday": "Morning",
        "Saturday": "Afternoon",
        "Sunday": "Evening"
    },

    "Bob": {
        "Monday": "Morning",
        "Tuesday": "Afternoon",
        "Wednesday": "Morning",
        "Thursday": "Evening",
        "Friday": "Afternoon",
        "Saturday": "Morning",
        "Sunday": "Evening"
    },

    "Charlie": {
        "Monday": "Evening",
        "Tuesday": "Morning",
        "Wednesday": "Afternoon",
        "Thursday": "Morning",
        "Friday": "Evening",
        "Saturday": "Afternoon",
        "Sunday": "Morning"
    },

    "David": {
        "Monday": "Afternoon",
        "Tuesday": "Evening",
        "Wednesday": "Morning",
        "Thursday": "Afternoon",
        "Friday": "Morning",
        "Saturday": "Evening",
        "Sunday": "Afternoon"
    },

    "Emma": {
        "Monday": "Morning",
        "Tuesday": "Morning",
        "Wednesday": "Evening",
        "Thursday": "Afternoon",
        "Friday": "Morning",
        "Saturday": "Morning",
        "Sunday": "Afternoon"
    },

    "Frank": {
        "Monday": "Evening",
        "Tuesday": "Afternoon",
        "Wednesday": "Evening",
        "Thursday": "Morning",
        "Friday": "Afternoon",
        "Saturday": "Morning",
        "Sunday": "Morning"
    },

    "Grace": {
        "Monday": "Afternoon",
        "Tuesday": "Morning",
        "Wednesday": "Morning",
        "Thursday": "Evening",
        "Friday": "Afternoon",
        "Saturday": "Evening",
        "Sunday": "Morning"
    }
}

# ---------------------------------------------------------
# Schedule Data Structure
# ---------------------------------------------------------

schedule = {}

for day in days:
    schedule[day] = {}
    for shift in shifts:
        schedule[day][shift] = []

# Track working days count
work_days = {}

for emp in employees:
    work_days[emp] = 0

# ---------------------------------------------------------
# Step 1: Assign preferred shifts
# ---------------------------------------------------------

MAX_EMPLOYEES_PER_SHIFT = 3
MAX_DAYS_PER_WEEK = 5

for emp, preferences in employees.items():

    for day, preferred_shift in preferences.items():

        # Employee already worked 5 days
        if work_days[emp] >= MAX_DAYS_PER_WEEK:
            continue

        # Check if employee already assigned today
        already_assigned = False

        for shift in shifts:
            if emp in schedule[day][shift]:
                already_assigned = True
                break

        if already_assigned:
            continue

        # Assign if preferred shift has space
        if len(schedule[day][preferred_shift]) < MAX_EMPLOYEES_PER_SHIFT:
            schedule[day][preferred_shift].append(emp)
            work_days[emp] += 1

        else:
            # -------------------------------------------------
            # Conflict Resolution
            # Assign another available shift same day
            # -------------------------------------------------
            assigned = False

            for shift in shifts:
                if shift != preferred_shift and len(schedule[day][shift]) < MAX_EMPLOYEES_PER_SHIFT:
                    schedule[day][shift].append(emp)
                    work_days[emp] += 1
                    assigned = True
                    break

            # -------------------------------------------------
            # If same day full, assign next available day
            # -------------------------------------------------
            if not assigned:
                current_index = days.index(day)

                for next_day in days[current_index + 1:]:

                    day_assigned = False

                    for shift in shifts:
                        if len(schedule[next_day][shift]) < MAX_EMPLOYEES_PER_SHIFT:

                            # ensure employee not already working that day
                            exists = False

                            for s in shifts:
                                if emp in schedule[next_day][s]:
                                    exists = True
                                    break

                            if not exists:
                                schedule[next_day][shift].append(emp)
                                work_days[emp] += 1
                                day_assigned = True
                                break

                    if day_assigned:
                        break

# ---------------------------------------------------------
# Step 2: Ensure minimum 2 employees per shift
# ---------------------------------------------------------

for day in days:

    for shift in shifts:

        while len(schedule[day][shift]) < 2:

            available = []

            for emp in employees:

                # employee should not exceed 5 days
                if work_days[emp] >= MAX_DAYS_PER_WEEK:
                    continue

                # employee should not already work today
                working_today = False

                for s in shifts:
                    if emp in schedule[day][s]:
                        working_today = True
                        break

                if not working_today:
                    available.append(emp)

            # randomly assign
            if len(available) > 0:
                chosen = random.choice(available)
                schedule[day][shift].append(chosen)
                work_days[chosen] += 1
            else:
                break

# ---------------------------------------------------------
# Final Weekly Schedule Output
# ---------------------------------------------------------

print("\n=================================================")
print("        FINAL EMPLOYEE WEEKLY SCHEDULE")
print("=================================================\n")

for day in days:

    print(f"\n{day}")
    print("-" * 40)

    for shift in shifts:

        print(f"{shift} Shift:")

        if len(schedule[day][shift]) == 0:
            print("  No Employees Assigned")

        else:
            for emp in schedule[day][shift]:
                print(f"  - {emp}")

    print()

# ---------------------------------------------------------
# Employee Work Summary
# ---------------------------------------------------------

print("\n=================================================")
print("           EMPLOYEE WORK SUMMARY")
print("=================================================\n")

for emp, count in work_days.items():
    print(f"{emp} works {count} day(s)")
```
